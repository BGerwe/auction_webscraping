```python
def extract_lots_table_from_soup(soup):
    soup = BeautifulSoup(response.data, 'html.parser')
    scripts = soup.find_all('script', type="text/javascript")
    table_raw = [script for script in scripts if 'lotModels' in script.text][0]
    table_dict = json.loads(table_raw.text.split('lotModels = ')[-1][:-4])
    return table_dict

def build_lots_df_from_table_dict(table_dict):
    lots_df = pd.DataFrame(table_dict)
    status_df = pd.DataFrame.from_dict(lots_df['lotStatus'].to_dict(), orient='index')
    lots_df = lots_df.merge(status_df, left_index=True, right_index=True)
    lots_df = lots_df[KEEP_COLS]
    lots_df = lots_df.rename(columns={'lead': 'item_name'})
    # First three lots are always their disclaimers
    lots_df = lots_df.loc[
        ~lots_df.item_name.isin(
            ['Mesa Public Auction NEW ENTRANCE STE 219',
             'Live Catalogue/ Soft Close STARTS 6pm every Sunday',
             'Pickup Location/ We do not ship! ENTER STE 219'
           ]
        )
    ]
    
    return lots_df
```

```python
base_url = 'https://mesapublicauction.hibid.com/lots/'#?apage=2&ipp=100'
keep_scraping = True
scraped_tables = []

for page_no in range(0, 20):
    if keep_scraping:
        print(f"Scraping data for page {page_no}")
        url = f"{base_url}?apage={page_no}&ipp=100"
        req = urllib3.PoolManager()
        response = req.request('GET', url)
        soup = BeautifulSoup(response.data, 'html.parser')
        scraped_table = extract_lots_table_from_soup(soup)
        print(len(scraped_table))
        if len(scraped_table):
            scraped_tables.append(scraped_table)
        else:
            print(f"No items scraped for page {page_no}")
            keep_scraping = False
```

```python
table_dicts = list(chain.from_iterable(scraped_tables))
lots_df = build_lots_df_from_table_dict(table_dicts)
```

```python
for col in lots_df.columns:
    if not col in ['item_name', 'lotNumber', 'eventItemId', 'itemId', 'timeLeftSeconds', 'timeLeftWithLimboSeconds', 'timeLeftTitle']:
        try:
            print(col, '---'*50)
            print(lots_df[col].unique())
        except:
            print(f"Can't print unique values for {col}")
```
